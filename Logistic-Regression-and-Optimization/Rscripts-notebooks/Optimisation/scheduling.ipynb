```python
import pandas as pd
import time
```

```python
import random

def generateData(machines=4, jobs=7, seed=0):
    PROC = [[0 for m in range(machines)] for j in range(jobs)]
    random.seed(seed)
    for j in range(jobs):
        for m in range(machines):
            PROC[j][m] = random.randint(1, 9)
    return PROC
```

```python
def greedy_algorithm(PROC):
    # Calculate sum of jobs, sum of time to finish job
    job_costs = [(i, sum(PROC[i])) for i in range(len(PROC))]
    # Greedy: Pick the fastest job, job can be finished fastest
    job_costs.sort(key=lambda x: x[1])
    # Return a sequence --> sequence of order --> index of the job in the matrix --> [3,2,1,0] --> job 3 run first then job 2,...
    return [job[0] for job in job_costs]
```

```python
def totalTime(PROC, seq):
    num_jobs = len(seq)
    num_machines = len(PROC[0])
    
    # Create a matrix to store the finish time of each machine
    # initial 0 matrix
    finish_times = [[0 for _ in range(num_machines)] for _ in range(num_jobs)]
    # matrix [m,j] has the same size as first matrix --> different values
    
    for i in range(num_jobs):
        job_idx = seq[i]
        for m in range(num_machines):
            if i == 0 and m == 0:
                # First job of first machine
                finish_times[i][m] = PROC[job_idx][m]
                # [0][0] = PROC[index of job][0]
                # for example the sequence return that job 3 is the first one --> PROC [3][0] --> first value of job 3
                # 3: [1,2,3,4] --> 1
                
            elif i == 0:
                # First job of other machines: wait the previous machine finish
                finish_times[i][m] = finish_times[i][m-1] + PROC[job_idx][m]
                # [1][m]= [1][m-1] + PROC[index of job][m]
                # for example first of job machine 2 --> finish_times[0][0] + PROC [3][1] --> second value of job 3
                # --> 3
            elif m == 0:
                # Job in the first machine: wait until machine 1 finish previous job
                finish_times[i][m] = finish_times[i-1][m] + PROC[job_idx][m]
                # job 2 is next --> [1][0] = finish_time[0][0] +PROC[2][0]
                # 2: [6,7,5,1] --> [1][0] = 1 + 6
                
            else:
                # Others:
                # Max (this job finish in before machine , this machine finish previous job)
                wait_for_job_to_finish_on_prev_machine = finish_times[i][m-1]
                wait_for_machine_to_be_free = finish_times[i-1][m]
                
                start_time = max(wait_for_job_to_finish_on_prev_machine, wait_for_machine_to_be_free)
                finish_times[i][m] = start_time + PROC[job_idx][m]
    
    # Final result
    return finish_times[num_jobs-1][num_machines-1]
```

```python
# change seed to 4 last digits of student ID
my_seed = 5197 
my_jobs = 100
my_machines = 10
PROC_DATA = generateData(machines=my_machines, jobs=my_jobs, seed=my_seed)

# Find the operated sequence by using greedy
greedy_seq = greedy_algorithm(PROC_DATA)

# Caculate total time
makespan = totalTime(PROC_DATA, greedy_seq)

print(f"machines and jobs: {PROC_DATA}")
print(f"Job order (Greedy): {greedy_seq}")
print(f"Total time (Makespan): {makespan}")
```
